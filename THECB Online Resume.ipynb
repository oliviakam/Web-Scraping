```python
##This code downloads 2022 4-YR CAREER ONLINE RESUMES

import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import os
from bs4 import BeautifulSoup

# List of User-Agent strings to mimic different browsers
user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:85.0) Gecko/20100101 Firefox/85.0",
    # Add more User-Agent strings as needed
]

# Define the URL you want to scrape
url = 'https://apps.highered.texas.gov/resumes/'

# Initialize a Selenium web driver with a random User-Agent
user_agent = random.choice(user_agents)
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(f"user-agent={user_agent}")
driver = webdriver.Chrome(options=chrome_options)

delay = random.uniform(3.0, 4.0)
time.sleep(delay)
    
driver.get(url)

select_element = driver.find_element(By.ID, "INSTTYPE")
dropdown = Select(select_element)
wait = WebDriverWait(driver, 10)
dropdown.select_by_value("CS4")

wait = WebDriverWait(driver, 10)

select_element = driver.find_element(By.ID, "FICE")
select_element.click()

wait = WebDriverWait(driver, 10)

dropdown = Select(select_element)

for _ in range(len(dropdown.options)):
    # Fetch the updated elements within the loop
    dropdown = Select(driver.find_element(By.ID, "FICE"))
    wait = WebDriverWait(driver, 10)
    fice_option = dropdown.options[_]
    fice_option_value = fice_option.get_attribute("value")
    fice_option_text = fice_option.text
    
    wait = WebDriverWait(driver, 30)
    dropdown.select_by_value(fice_option_value)
    
    wait = WebDriverWait(driver, 10)
    #year_select = driver.find_element(By.ID, "YEAR_TEXT")
    #year_dropdown = Select(year_select)
    #year_dropdown.select_by_value('2021')
    
    
    submit_button = driver.find_element(By.ID, "Submit")
    submit_button.click()
    
    driver.switch_to.window(driver.window_handles[-1])
    pdf_url = driver.current_url
    response = requests.get(pdf_url)
    
    year = '2022'
    file_name = f"Online Resumes {year}{fice_option_value}PUB.pdf"
    file_path = f"/Users/oliviaenriquez/Desktop/pdfs/4-Yr Career/2022/{file_name}"
    
    with open(file_path, "wb") as pdf_file:
        pdf_file.write(response.content)
    print(f"Online Resume for {fice_option_text} with FICE numer {fice_option_value} successfully downloaded")
    
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    wait = WebDriverWait(driver, 10)

    select_element = driver.find_element(By.ID, "FICE")
    select_element.click()

    wait = WebDriverWait(driver, 10)
```
